# 3. データの前処理（数値化、分布の確認、欠損値の確認）
- ラベルはone-hot-codingする
- ヒストグラムの表示

```python
i = 0
plt.figure(figsize=[24, 15])
for i in range(df_train.columns.shape[0]):
    plt.subplot(4, 3, i+1)
    plt.title(df_train.columns[i])
    if df_train.iloc[:, i].dtype == ('int64' or 'float64'):
        plt.hist(np.array(df_train.iloc[:, i]), bins=21, rwidth=0.8)
    else:
        df_train.iloc[:, i].value_counts().plot(kind='bar')
plt.tight_layout()
plt.show()
```

- "Name", "PassengerId"は分布がないので削除しても良さそう
- （Nameに関しては、国籍を推測することで判断材料につながることも苗字が同じであれば子供の年齢を推定することもできそうだけれども、今回は割愛）

### "Age", "PassengerId"の削除

```python
df_train.drop(["Name", "PassengerId"], axis=1, inplace=True)
df_test.drop(["Name", "PassengerId"], axis=1, inplace=True)
```

```python
plt.figure(figsize=[24, 15])
for i in range(df_train.columns.shape[0]):
    plt.subplot(4, 3, i+1)
    plt.title(df_train.columns[i])
    if df_train.iloc[:, i].dtype == ('int64' or 'float64'):
        plt.hist(np.array(df_train.iloc[:, i]), bins=21, rwidth=0.8)
    else:
        df_train.iloc[:, i].value_counts().plot(kind='bar')
plt.tight_layout()
plt.show()
```

### Pivotingによる特徴量調査

- Pclassと生存者の相関関係

```python
for i in range(df_train.columns.shape[0]):
    current_column = df_train.columns[i]
    if current_column == "Survived":
        continue
    if df_train[current_column].nunique() > 10:
        continue
    print(df_train[[current_column, 'Survived']].groupby([current_column], as_index=False).mean().sort_values(by='Survived', ascending=False))
    print('-'*40)
```

```python
g = sns.FacetGrid(df_train, col='Survived', size=4)
g.map(plt.hist, 'Age', bins=21, rwidth=0.8, color='blue')
```

```python
# grid = sns.FacetGrid(train_df, col='Pclass', hue='Survived')
grid = sns.FacetGrid(df_train, col='Survived', row='Pclass', size=2.4, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend();
```
